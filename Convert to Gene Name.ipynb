```python
def convert_name(dic, gene_dic):
    new_dic = dict()
    for key,value in dic.items():
        if key[0] in gene_dic:
            key = (gene_dic.get(key[0]), key[1])
        if key[1] in gene_dic:
            key = (key[0], gene_dic.get(key[1]))
        new_dic[key] = value
    return new_dic
```

```python
y = convert_name(x, w)
```
