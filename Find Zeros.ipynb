```python
#BROKEN FUNCTION SCROLL DOWN FOR UPDATED ONE
def update_zero1(dic):
    new_dic = dic
    check_list = []
    del_list = []
    for key, value in new_dic.items():
        count = 0
        for element in value:
            if element == 0:
                if count == 0:
                    if (abs((value[2] - value[1]) - (value[4] - value[3])) < 0.2):
                        value[0] = value[1] - (value[2] - value[1])
                        check_list.append([key, value, 0])
                        break
                    else:
                        del_list.append([key, value])
                        break
                if count == 1:
                    if (abs((value[2] - value[0]) - ((value[4] - value[3]) * 2)) < 0.2):
                        value[1] = value[2] - (value[4] - value[3])
                        check_list.append([key, value, 1])
                        break
                    else:
                        del_list.append([key, value])
                        break
                if count == 2:
                    if (abs((value[1] - value[0]) - (value[4] - value[3])) < 0.2):
                        value[2] = value[1] + (value[1] - value[0])
                        check_list.append([key, value, 2])
                        break
                    else:
                        del_list.append([key, value])
                        break
                if count == 3:
                    if (abs(((value[1] - value[0]) * 2) - (value[4] - value[2])) < 0.2):
                        value[3] = value[4] - (value[2] - value[1])
                        check_list.append([key, value, 3])
                        break
                    else:
                        del_list.append([key, value])
                        break
                if count == 4:
                    if (abs((value[1] - value[0]) - (value[3] - value[2])) < 0.2):
                        value[4] = value[3] + (value[1] - value[0])
                        check_list.append([key, value, 4])
                        break
                    else:
                        del_list.append([key, value])
                        break
            count = count + 1
                
    
    print(len(check_list))
    print(len(del_list))
    #print(del_list)
    print(check_list)
    return new_dic
```

```python
def predict_zeros(dic):
    new_dic = dic
    dicList = []
    del_list = []
    for key, value in new_dic.items():
        if 0 in value:
            loc = value.index(0)
            value.remove(0)
            if loc in [0,2,4]:
                if abs((value[1] - value[0]) - (value[3] - value[2])) < 0.1:
                    if loc == 0:
                        if (((value[0] - (value[1]-value[0])) < 1) and ((value[0] - (value[1]-value[0])) > 0)):
                            value.insert(loc, value[0] - (value[1]-value[0]))
                            dicList.append([key, value, loc])
                        else:
                            del_list.append(key)
                    if loc == 2:
                        if (((value[2] - (value[3]-value[2])) < 1) and ((value[2] - (value[3]-value[2])) > 0)):
                            value.insert(loc, value[2] - (value[3]-value[2]))
                            dicList.append([key, value, loc])
                        else:
                            del_list.append(key)
                    if loc == 4:
                        if (((value[3] + (value[3]-value[2])) < 1) and ((value[3] + (value[3]-value[2])) > 0)):
                            value.insert(loc, value[3] + (value[3]-value[2]))
                            dicList.append([key, value, loc])
                        else:
                            del_list.append(key)
                else:
                    del_list.append(key)
            elif loc in [1]:
                if abs((value[1] - value[0]) - ((value[3] - value[2])*2)) < 0.1:
                    if (((value[1] - (value[3]-value[2])) < 1) and ((value[1] - (value[3]-value[2])) > 0)):
                        value.insert(loc, value[1] - (value[3]-value[2]))
                        dicList.append([key, value, loc])
                    else:
                        del_list.append(key)
                else:
                    del_list.append(key)
            elif loc in [3]:
                if abs(((value[1] - value[0])*2) - (value[3] - value[2])) < 0.1:
                    if (((value[3] - (value[1]-value[0])) < 1) and ((value[3] - (value[1]-value[0])) > 0)):
                        value.insert(loc, value[3] - (value[1]-value[0]))
                        dicList.append([key, value, loc])
                    else:
                        del_list.append(key)
                else:
                    del_list.append(key)
        
    for element in del_list:
        del new_dic[element]             
    
    print(dicList)
    print(len(dicList))
    return new_dic
```
